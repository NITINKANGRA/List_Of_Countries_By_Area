# **Here I Have Changed The Datatype**

```python
data['% water'] = pd.to_numeric(data['% water'], errors='coerce')
```

```python
data.info()
```

# **while changing the datatype of 'rank' column i.e object to int . first changes it to numeric**

```python
data['Rank'] = pd.to_numeric(data['Rank'], errors='coerce')
data = data.dropna(subset=['Rank'])
data['Rank'] = data['Rank'].astype(int)
```

```python
data
```

```python
data.info()
```

# **Here first I replaces comma(,) into decimal(.) makes it to float , then changes to int.**

```python
data['Total in km2'] = data['Total in km2'].str.replace(',', '')
data['Total in km2'] = data['Total in km2'].astype(float)
data['Total in km2'] = data['Total in km2'].astype(int)
```

```python
data['Land in km2'] = data['Land in km2'].str.replace(',', '')
data['Land in km2'] = data['Land in km2'].astype(float)
data['Land in km2'] = data['Land in km2'].astype(int)
```

```python
data['Water in km2'] = pd.to_numeric(data['Water in km2'], errors='coerce')
data['Water in km2'] = data['Water in km2'].fillna(0).astype(int)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data
```
